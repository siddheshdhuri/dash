```python
input_csv_dir = '../input/csv'
input_pickle_dir = '../input/pickle'

for path, subdirs, files in os.walk(input_csv_dir):
    for name in files:
        df_name = name.split('.')[1] #get df name from filename
        df = pd.read_csv(os.path.join(path, name))        
        df.to_pickle(os.path.join(input_pickle_dir, df_name))        
```

#### ET_LOCMAT

Definition: Contains the Initial Stock present for the product at the locations. Also have the information about the Safety Stock Violation Cost (Per EA Per Day) and Storage Cost (Per EA Per Day) at each location

BUCFL: Ind.: Storage Costs per Day Storage Costs/Day <br>
CONIN: Handling Capacity Consumption (GR) <br>
CONOU: Handling Capacity Consumption (Gds Iss.) <br>
COSIN: Handling Costs (Goods Receipt) <br>
HCOST: Storage Costs Per Prod. <br>
LOCID: Location Id <br>
MATID: Material Id <br>
MAXFL: Indicator: Storage capacity active <br>
MAXST: Material Storage Capacity <br>
SSPEN: Penalty for not enough safety stock <br>
STOCK: Initial Warehouse Stock <br>
SUBID: SublocatID <br>

#### ET_LOCPROD

Definition: Contains the In-Transit and the safety Stock Quantity and also the Safety Stock number of days maintained at the location

FDIST: Aggregated Distribution Quantity <br>
FPROD: Fixed Production (Including In-Transits)<br>
FSUBST: Aggregated Substitution Quantity <br>
LOCID: location id <br>
MATID: material id <br> 
MAXST: Material Storage Capacity <br>
SAFTY: Safety Stock <br>
SAPEN: Penalty for not enough safety stock <br>
STPEN: Penalty for exceeding stock on hand <br>
SVTTY: Safety Days' Supply <br>
